```python
import pandas as pd
```

```python
df = pd.read_csv("homep.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
merged = merged.drop(['town','west windsor'],axis=1)
merged
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = merged.drop('price',axis=1)
x
```

```python
y = merged.price
```

```python
model.fit(x,y)
```

```python
y_pred = model.predict([[2800,0,1]])
y_pred
```

```python
model.predict([[2500,0,0]])
```

```python
model.score(x,y)*100
```
