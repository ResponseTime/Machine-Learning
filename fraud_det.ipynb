```python
from sklearn.linear_model import LogisticRegression
```

```python
df.head(10)
```

```python
fraud = df['Class'].value_counts()
```

```python
fraud
```

```python
import matplotlib.pyplot as plt
fraud_percentage = fraud / len(df) * 100

plt.figure(figsize=(8, 6))
fraud.plot(kind='bar', color=['blue', 'red'])
plt.title('Number of Fraudulent vs Non-Fraudulent Transactions')
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
plt.xticks([0, 1], labels=['Non-Fraudulent', 'Fraudulent'], rotation=0)
plt.show()

print("Fraudulent Transactions:")
print(f"Count: {fraud[1]}")
print(f"Percentage: {fraud_percentage[1]:.2f}%")

print("\nNon-Fraudulent Transactions:")
print(f"Count: {fraud[0]}")
print(f"Percentage: {fraud_percentage[0]:.2f}%")
```

```python
x = df.copy()
y = x.pop('Class')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
X_scaled = Scaler.fit_transform(X_train)
X_test_scaled = Scaler.transform(X_test)
```

```python
model = LogisticRegression(random_state=50)
model.fit(X_scaled, y_train)
```

```python
model.score(X_test_scaled,y_test)*100
```

```python
import seaborn as sn
from sklearn.metrics import confusion_matrix
```

```python
plt.figure()
y_pred = model.predict(X_test_scaled)
cm = confusion_matrix(y_test,y_pred)
sn.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
```
